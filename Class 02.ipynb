The code below doesn't work as expected. Can you figure out what is going wrong and how to make it work? We want to calculate the volatility of daily returns (not prices) within each year.

```python
sp500.pct_change().dropna().groupby('yr')['Close'].std()
```

```python
# a very nice alternative solution that Nick led us to with great question, and 
# then Akash helped us find

sp500['Close'].pct_change().groupby(sp500.index.year).std()
```

```python
sp500.pct_change().dropna().groupby('yr')['Close'].std()
```

```python
# Also not a solution...

sp500.groupby('yr')['Close'].pct_change().std()
```

```python
# this works... try to understand why and we'll discuss in class

sp500.groupby('yr')['Close'].apply(lambda x: x.pct_change().std())
```

```python
def max_minus_min(x):
  return x.max() - x.min()
```

```python
max_minus_min(np.array([3, -7, 12, 4]))
```

```python
sp500.groupby('yr')['Adj Close'].apply(max_minus_min).plot();
```

Now we're going to group by a variable that isn't actually in our data. Above, we grouped by the values in the `yr` column. We want to group by the month associated with the date, and this isn't a column.

Sure, we could create a column that contains the month, but that would add another unneccesary step. Instead, we simply pass a list of values that contain the month.

```python
# construct a return series to work with
spret = sp500['Adj Close'].pct_change()
```

```python
spret.groupby(spret.index.month).mean().plot.bar();
```

OK, I lied. We're not passing a `list`. It's actually an `Index` object, which is [list-like](https://pandas.pydata.org/docs/reference/api/pandas.api.types.is_list_like.html) meaning it's, um, like a list.

```python
spret.index.month
```

```python
from pandas.api.types import is_list_like

is_list_like(spret.index.month)
```

We can apply an arbitrary function to grouped data.

```python
def tstat(x):
  return x.mean() / (x.std() / np.sqrt(len(x)))
```

```python
spret.groupby(spret.index.month).apply(tstat)
```

```python
# use a lambda function to do the same thing

spret.groupby(spret.index.month).apply(lambda x: x.mean()/(x.std()/np.sqrt(len(x))))
```
