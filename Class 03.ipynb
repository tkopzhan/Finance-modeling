```python
dowrets = {}
for ticker in djia.index:
  dowrets[ticker] = pdr.get_data_yahoo(ticker)['Adj Close'].pct_change()
```

```python
dowrets = pd.DataFrame(dowrets).dropna()
```

```python
dowrets.info()
```

```python
# Do the same thing with less code using a dictionary comprehension

dowrets = pd.DataFrame(
    {ticker:pdr.get_data_yahoo(ticker)['Adj Close'].pct_change() 
      for ticker in djia.index}
    )
```

```python
dowrets = dowrets.dropna(how='all').iloc[:-1]
```
